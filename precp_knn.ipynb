```python
atp = pd.read_table('precipitationwar.xls')
```

```python
s_months = list(atp.columns)
```

```python
s_months.remove('Year')
```

```python
months = []
for i in range(12):
    months.append(i+1)
    
months
```

```python
m_map = {}

for i in range(12):
    m_map[s_months[i]] = i+1
    
m_map
```

```python
avgscores = []
```

```python
atpr = atp.drop('Year', axis=1)
```

```python
np_df = atpr.as_matrix()
```

```python
np_df
```

```python


for i in range(102):
    for k in range(12):
        avgscores.append(np_df[i,k])
```

```python
avgscores
```

```python
months = months*102
months
```

```python
year = atp['Year']
```

```python
years = list(year)
```

```python
k = []
```

```python
for year in years:
    for i in range(12):
        k.append(year)
```

```python
len(k)
```

```python
df = pd.DataFrame(
    {'year': k,
     'months': months,
     'precp': avgscores
    })
```

```python
cols = list(df.columns)
```

```python
cols
```

```python
cols.remove('precp')
cols.remove('year')
```

```python
X = df[cols]
Y = df['precp']
type(X)
```

```python
train_X= X[:864]
train_Y = Y[:864]
```

```python
print(train_X.shape)
print(train_Y.shape)
```

```python
test_X = X[-360:]
test_Y = Y[-360:]
```

```python
print(test_X.shape)
print(test_Y.shape)
```

```python
from sklearn.neighbors import KNeighborsRegressor
neigh = KNeighborsRegressor(n_neighbors=2)
```

```python
k = list(range(26))
k.remove(0)
for i in k:
    neigh = KNeighborsRegressor(n_neighbors=i)
    neigh.fit(train_X, train_Y)
    print(" train on {} = ".format(i)+ str(neigh.score(train_X,train_Y)))
    print(" test on {} =  ".format(i) + str(neigh.score(test_X,test_Y)))
```

```python
#the best k is 11 which can be inferred from above scores

neigh = KNeighborsRegressor(n_neighbors=11)
neigh.fit(train_X, train_Y)
print(neigh.score(train_X,train_Y))
print(neigh.score(test_X,test_Y))
```
